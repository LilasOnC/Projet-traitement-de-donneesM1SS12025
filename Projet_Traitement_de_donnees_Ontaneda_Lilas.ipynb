**Remplacement des valeurs manquantes**

Pour pouvoir utiliser 'begin' et 'end' pour analyser la période on doit compléter les données manquantes au mieux. On peut utiliser le temps moyen entre 'begin' et 'end' des groupes qui ont ces deux données pour compléter ceux qui n'ont que 'begin' ou de 'end'. L'ajout du temps moyen de vie d'un groupe peut dépasser la date d'aujourd'hui, on les borne donc à 2024.

```python
#Transforme les données en valeurs numériques
df_artists['begin'] = pd.to_numeric(df_artists['begin'], errors='coerce').astype('Int64')
df_artists['end'] = pd.to_numeric(df_artists['end'], errors='coerce').astype('Int64')

#Moyenne du temps d'activité/de vie des groupes ayant les données begin et end, l'arrondi parce que c'est una année
mean_life_artist = round((df_artists['end'] - df_artists['begin']).mean())
# mean_life_artist

#Sous partie avec les artistes qui ont un begin mais pas de end
to_change_begin = df_artists.loc[df_artists['begin'].isnull() & df_artists['end'].notnull()]

#Sous partie avec les artistes qui ont un end mais pas de begin
to_change_end = df_artists.loc[df_artists['end'].isnull() & df_artists['begin'].notnull()]

#Modifie le df pour remplacer les valeurs manquantes
df_artists.loc[to_change_begin.index, 'begin'] = df_artists.loc[to_change_begin.index, 'end'] - mean_life_artist
df_artists.loc[to_change_end.index, 'end'] = df_artists.loc[to_change_end.index, 'begin'] + mean_life_artist

#Borne à 2024
df_artists.loc[df_artists['end'] >= 2024, 'end'] = 2024
```

```python
min_begin = df_artists['begin'].min()
df_artists.loc[df_artists['begin'] == min_begin]
```
